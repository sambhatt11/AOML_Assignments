```python
import pandas as pd
import sklearn
print(sklearn.__version__)
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
df['BMI_Category'] = df['BMI'].apply(
    lambda bmi: 'Underweight' if bmi < 18.5 else
                'Normal weight' if 18.5 <= bmi < 22.9 else
                'Overweight' if 23.0 <= bmi < 27.4 else
                'Obese' if 27.5 <= bmi < 32.9 else
                'Severely Obese'
)
```

```python
df
```

```python
df.info()
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_test = train_test_split(X,y, test_size=0.2)
X_train
```

```python
X_train['BMI_Category'] = X_train['BMI_Category'].astype(str)
X_val['BMI_Category'] = X_val['BMI_Category'].astype(str)
```

```python
from sklearn.preprocessing import StandardScaler
numeric_features = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]
scaler = StandardScaler()
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_val[numeric_features] = scaler.transform(X_val[numeric_features])
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")
X_train_encoded = pd.DataFrame(encoder.fit_transform(X_train[["BMI_Category"]]),
                               columns=encoder.get_feature_names_out(["BMI_Category"]),
                               index=X_train.index)
X_val_encoded = pd.DataFrame(encoder.transform(X_val[["BMI_Category"]]),
                             columns=encoder.get_feature_names_out(["BMI_Category"]),
                             index=X_val.index)
```

```python
X_train = pd.concat([X_train[numeric_features], X_train_encoded], axis=1)
X_val = pd.concat([X_val[numeric_features], X_val_encoded], axis=1)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_val)
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=7)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_val)
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
joblib.dump(encoder, "encoder.pkl")
joblib.dump(dt, "model.pkl")
```

```python
def inference_pipeline(sample):
  scaler = joblib.load("scaler.pkl")
  encoder = joblib.load("encoder.pkl")
  knn = joblib.load("model.pkl")
  sample_numeric = df[numeric_features]
  sample_categorical = df[['BMI_Category']]
  sample_numeric_scaled = scaler.transform(sample_numeric)
  sample_categorical_encoded = encoder.transform(sample_categorical)
  sample_processed = pd.DataFrame(sample_numeric_scaled, columns=numeric_features)
  sample_categorical_encoded = pd.DataFrame(encoder.transform(sample_categorical),
                                        columns=encoder.get_feature_names_out(["BMI_Category"]),
                                        index=sample_categorical.index) # get feature names directly from encoder
  sample_final = pd.concat([sample_processed.reset_index(drop=True), sample_categorical_encoded.reset_index(drop=True)], axis=1)
  prediction = dt.predict(sample_final)
  return prediction
```
